### 2. Select variables from data set suitable for exploratory visual analysis

```python
#creating subset with selected variables
happiness_clean = happiness_wrangled[['Happiness Score','Economy (GDP per Capita)','Social Support', 'Health (Life Expectancy)', 'Freedom to make life choices', 'Lack of Corruption', 'Donations to Charity']]
```

### 3. Correlation Matrix Heatmap

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(happiness_clean.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to place the correlation coefficients onto the heatmap
plt.savefig('result.png', bbox_inches='tight', pad_inches=0.0) #save image
```

### Relationships between the variables according to matrix above

i) 0.79 - Strongest correlation is between Happiness Score and Economy(GDP per Capita)- The Wealthier the country the          Happier the people are

ii) 0.74 - Strong Correlation between Happiness Score and Life Expectancy- Countries with high life expectancy are happier

ii) 0.65 - Strong correlation between Happiness Score and Social support - Countries with stronger social networks are happier

iii) 0.55 - Strong correlation between Happiness Score and  Freedom - Countries whose people can make life choices are happier

iv)0.26 - Weak correlation between Happiness Score and Donations to Charity - Generosity has negligible effect on happiness

v)0.23 - Weak correlation between Happiness and Lack of corruption - Corruption or the lack of it, has minimal effect on Happiness

vi) 0.79 - Strong correlation between Economy and Life Expectancy - The stronger the Economy the Higher the Life Expectancy and vice versa

vii) 0.59 - Strong correlation between Economy and Social support - The stronger the Economy the stronger the social support and vice versa

viii) 0.58 - Strong Correlation between Life Expectancy and Social Support - The stronger the social support the higher the life expectancy and vice versa

ix) 0.43 - Moderate correlation between Social Support and  Freedom -Having Freedom does not necessarily indicate Social support 

x) 0.36 - Moderate correlation between Freedom and lack of corruption - Freedom to make life choices does not necessarily mean the country is free of corruption or vice versa

xi) 0.34 - Moderate correlation between Freedom and life Expectancy -  Having Freedom does not always indicate high Life expectancy

xii) 0.33 - Moderate correlation between Freedom to make life choices and Generosity - average connection between Freedom to make life choices and level of Generosity

All other correlations indicate weak or negative relationships and do not need to be analyzed further


### 4. Scatterplot of Strongest Correlation

```python
# Create a scatterplot for the "Happiness Score" and "Economy(GDP per Capita)" columns in seaborn

sns.lmplot(x = 'Happiness Score', y = 'Economy (GDP per Capita)', data = happiness_clean)
```

#### The Scatter plot above shows a positive, linear, strong relationship between Economy(GDP per Capita) of a country and its level of Happiness. Generally, this indicates that the wealthier the country, the happier its residents. There are a few outliers majority of which indicate some countries with weak economies showed above average level of Happiness.

### 6. Categorical plot 

```python
# Use a histogram to visualize the distribution of the variables. 

sns.histplot(happiness_clean['Happiness Score'], bins = 20, kde = True) # The argument "kde" add a line that encompasses the distribution
plt.savefig('Categorical_plot.png', bbox_inches='tight', pad_inches=0.0) #save image
```

#### Create Happiness Categories

```python
happiness_clean.loc[happiness_clean['Happiness Score'] <4, 'Happiness Category'] = 'Not Happy'
```

```python
happiness_clean.loc[(happiness_clean['Happiness Score'] >=4) & (happiness_clean['Happiness Score'] < 6), 'Happiness Category'] = 'Moderately Happy'
```

```python
happiness_clean.loc[(happiness_clean['Happiness Score'] >=6) & (happiness_clean['Happiness Score'] <=8), 'Happiness Category'] = 'Happy'
```

```python
# Happiness category count
happiness_clean['Happiness Category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the Happiness Categories

sns.set(style="ticks")
g = sns.catplot(x="Economy (GDP per Capita)", y="Happiness Score", hue="Happiness Category", data=happiness_clean)
plt.savefig('Categories_scatter.png', bbox_inches='tight', pad_inches=0.0) #save image
```

```python
#Save image
g.savefig("out.png")
```

#### The categorical plot above shows that Countries with higher GDP have a higher Happiness Score
