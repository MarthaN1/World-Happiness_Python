Countries with more than 2 years data missing to be dropped and excluded from analysis
Djibouti, Gambia, Oman, Puerto Rico, Somaliland Region, Suriname, Swaziland


```python
#Create subset that excludes Countries missing more than 2 years data 
df_happiness_deletion = df_happiness.loc[df_happiness['Country'].isin(['Djibouti','Gambia', 'Oman', 'Puerto Rico', 'Somaliland region','Somaliland Region', 'Suriname', 'Swaziland'])== False]
```

Missing Values

```python
 #check if there are any missing values
df_happiness_deletion.isnull().sum()
```

```python
# checking for median value
df_happiness_deletion.median()
```

Imputing missing values with median values

```python
#imputing median value(5.353) for missing Happiness Score 
df_happiness_deletion['Happiness Score'].fillna(5.323, inplace=True) 
```

```python
#imputing median value(0.9825) for missing Economy (GDP per Capita) 
df_happiness_deletion['Economy (GDP per Capita)'].fillna(0.9825, inplace=True)
```

```python
#imputing median value(1.1265) for missing Social Support 
df_happiness_deletion['Social Support'].fillna(1.1265, inplace=True)
```

```python
#imputing median value(0.6505) for missing Health (Life Expectancy)
df_happiness_deletion['Health (Life Expectancy)'].fillna(0.6505, inplace=True)
```

```python
#imputing median value(0.4310) for missing Freedom to make life choices
df_happiness_deletion['Freedom to make life choices'].fillna(0.4310, inplace=True)
```

```python
#imputing median value(0.1550) for missing Lack of Corruption
df_happiness_deletion['Lack of Corruption'].fillna(0.1550, inplace=True)
```

```python
#imputing median value(0.1270) for missing Donations to Charity
df_happiness_deletion['Donations to Charity'].fillna(0.1270, inplace=True)
```

```python
#imputing value(0) for missing Happiness Rank
df_happiness_deletion['Happiness Rank'].fillna(0, inplace=True)
```
